```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Data/PopFixed.csv", delimiter=",", thousands=",", decimal=".", index_col="Country (or dependency)")
df.drop(columns=["#"], inplace=True)
```

```python
df.plot()  # Pandas hat eine direkte Integration mit matplotlib
```

```python
df["Population(2020)"].plot()
```

## Zwei Modi

Globaler Modus und Objektorientierter Modus

```python
plt.figure(figsize=(12, 6))
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(df["Population(2020)"])  # Global
```
