## Lineare Regression

Zukunftswerte vorhersagen anhand von vergangenen Werten

```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
df = pd.read_csv("Data/SeoulBikeData.csv")
```

```python
df
```

```python
df.drop(columns=["Seasons", "Holiday", "Functioning Day"], inplace=True)
```

```python
df
```

Weitere unnütze Spalten:
- Datum
- Stunde nur 12 Uhr (dauert sonst zu lange)

```python
df.drop(columns=["Date"], inplace=True)
```

```python
df = df[df["Hour"] == 12]
```

```python
df
```

```python
df.drop(columns=["Visibility (10m)", "Rainfall(mm)", "Snowfall (cm)"], inplace=True)
```

```python
df
```

```python
df.drop(columns=["Hour"], inplace=True)
```

```python
df = df.rename(columns={"Rented Bike Count": "bike_count", "Temperature(°C)": "temp", "Humidity(%)": "humidity", "Wind speed (m/s)": "wind", "Dew point temperature(°C)": "dew", "Solar Radiation (MJ/m2)": "radiation"})
```

```python
df
```

```python
sample = df.sample(frac=1)  # Mischt das Datenset
sample
```

```python
gesamt = len(df)
trainingLen = int(gesamt * 0.6)
validLen = int(gesamt * 0.2) + trainingLen
```

```python
training, valid, test = np.split(sample, [trainingLen, validLen])
```

```python
len(training)
```

```python
len(valid)
```

```python
len(test)
```

```python
def scaleDataset(data):
    from sklearn.preprocessing import StandardScaler
    left = data[data.columns[0]].values.reshape(-1, 1)
    temp = data[data.columns[1]].values
    scaler = StandardScaler()
    return scaler.fit_transform(left), temp.reshape(-1, 1)
```

```python
train_bike_count, train_temp = scaleDataset(training)
valid_bike_count, valid_temp = scaleDataset(valid)
test_bike_count, test_temp = scaleDataset(test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(train_bike_count, train_temp)
```

```python
reg.score(test_bike_count, test_temp)
```

```python
pred_temp = reg.predict(test_bike_count)
```

```python
comparison = pd.DataFrame(np.hstack((pred_temp, test_temp)))
```

```python
comparison["%1"] = comparison[0] / comparison[1]
comparison["%2"] = comparison[1] / comparison[0]
```

```python
comparison
```

## Tensorflow

Normalizer: Entfernt Outlier

```python
tf.keras.layers.Normalization(input_shape=(1,), axis=0)
```

```python
model = tf.keras.Sequential([
        tf.keras.layers.Normalization(input_shape=(1,), axis=None),
        tf.keras.layers.Dense(64, activation="relu"),
        tf.keras.layers.Dropout(0.5),
        tf.keras.layers.Dense(64, activation="relu"),
        tf.keras.layers.Dropout(0.5),
        tf.keras.layers.Dense(64, activation="relu"),
        tf.keras.layers.Dense(1, activation="sigmoid")
    ])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(0.001), loss="mean_squared_error")
```

```python
history = model.fit(
    train_bike_count, train_temp,
    verbose=1,
    epochs=500,
    validation_data=(valid_bike_count, valid_temp))
```

```python
model.predict(test_bike_count)
```
