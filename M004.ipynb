## Andere Fehler ersetzen

z.B. Prozente, N.A.

```python
df["Fert.Rate"] == "N.A."
```

```python
df[df["Fert.Rate"] == "N.A."]
```

```python
df.loc[df["Fert.Rate"] == "N.A.", "Fert.Rate"] = 0
```

```python
df
```

```python
df.loc[df["Med.Age"] == "N.A.", "Med.Age"] = 0
```

```python
df
```

```python
def fixNA(df, column, newValue):
    df.loc[df[column] == "N.A.", column] = newValue
```

```python
fixNA(df, "UrbanPop %", 0)
```

```python
df
```

```python
df["YearlyChange"] = df["YearlyChange"].replace(" %", "", regex=True)
```

```python
df
```

```python
def fixPct(df, column, newValue):
    df[column].replace(" %", newValue, regex=True, inplace=True)
```

```python
fixPct(df, "UrbanPop %", "")
```

```python
fixPct(df, "WorldShare", "")
```

```python
df
```

```python
df.info()
```

```python
df = df.convert_dtypes()
```

```python
df.info()
```

```python
df["YearlyChange"] = df["YearlyChange"].astype(float)
df["Fert.Rate"] = df["Fert.Rate"].astype(float)
df["Med.Age"] = df["Med.Age"].astype(int)
df["UrbanPop %"] = df["UrbanPop %"].astype(int)
df["WorldShare"] = df["WorldShare"].astype(float)
```
